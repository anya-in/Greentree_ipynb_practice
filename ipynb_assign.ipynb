This notebook demonstrates how to read a CSV file using `pandas` and plot a simple graph using `matplotlib`.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("predicted_delays.csv")
df.head()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df["FlightNum"], df["Cancelled"], marker='o', linestyle='-', color='skyblue')
plt.title("Most commonly cancelled Flights")
plt.xlabel("FlightNum")
plt.ylabel("Cancelled")
plt.grid(True)
plt.tight_layout()
plt.show()
```
